```python
''' In this task, we want to explore possible correlations among variables, respectively, data points. To make statements
about presence of absence of any kind of relationship we can express this both
in a quantitive way (computation of a correlation measure) and in a qualitative way, i.e.
using data visualization. '''

# Computation of correlation coefficient. 
 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

np.random.seed(1)
```

```python
# 1000 random integers between 0 and 50

x_data = np.random.randint(0, 50, 1000)
y_data = x_data + np.random.normal(0, 10, 1000)

# Is the correlation positive, negative or is there no correlation?

print np.corrcoef(x_data, y_data)
```

#### 1._Which visualization you can use to display your data and check for possible relationships?_

The plot-function of the matplotlib.pyplot class can be used to view and check for a (linear) relationenship between two variables (features). The plot-function - as used here - can take two lists with corresponding x and y values for Points in a Cartesian coordinate-system and String signifying the way to format the output.


```python
# Which visualization to use here you learnt in lecture 3

plt.plot(x_data, y_data, 'm1')
# x and y are strongly positively correlated.
```

#### 2. *Detected correlation for x_data,y_data:*

x and y are (strongly) positively correlated.
The pearson coefficient is positive and quite high:

```python
def pcc(x,y):
    meanx = x.mean()
    meany = y.mean()
    stdx = x.std()
    stdy = y.std()
    return ((x-meanx)*(y-meany)).mean()/(stdx*stdy)
  # return ((x*y).mean()-meanx*meany)/(stdx*stdy)

print pcc(x_data,y_data)
```

```python
# Let us try another example

x_data2 = np.random.randint(0, 50, 1000)
y_data2 = 100 - x_data2 + np.random.normal(0, 5, 1000)
print np.corrcoef(x_data2, y_data2)
```

```python
plt.plot(x_data2, y_data2, 'y>')
# x and y are strongly negatively correlated.
```

x and y are strongly negatively correlated.
