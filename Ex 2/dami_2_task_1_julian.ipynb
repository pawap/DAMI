```python
data = loadtxt('apartment_prices.dat')

plt.ylabel('price in euros')
plt.title('Apartment prices')
plt.legend()

plt.boxplot(data)

# The boxplot allows us to get a quick overview about some important characteristic numbers from the data:
# 1. We get to see the median of the data. This is the colored line within the box.
# 2. We get to see the range for the whole data between the two caps at the top end and at the lower end --
# except for outliers which are shown as single dots above or below this range.
# 3. We get to see the range for the bulk of our data as we know that 50 % of our data is within the box.
# 4. We get to see all quartiles which are the top and the lower end of the box together with the median line,
# so we also know that there is 25 % of the data in each between the median line and the top end of the box and
# between the median line and the lower end of the box and also between the top end of the box and the top cap
# as well as between the lower end of the box and the lower cap.

# For this specific data we thought the boxplot would be the best fit as we want to get an estimation about the
# "average" price for an apartment. You care about the price range that seems acceptable and want to know at
# which price an apartment is probably too expensive. You don't care about the exact numbers though --
# how many apartments there are at a specific price is not that interesting. The boxplot allows us to get exactly
# the information we need as it doesn't show how many apartments there are for a specific price or a range of
# prices but instead gives us the meta information we want.
```

```python
data = loadtxt('grades_lecture1.dat')
 
plt.xlabel('grade')
plt.ylabel('number of students')
plt.title('Grades lecture 1')
plt.legend()

plt.hist(data,(1,2,3,4,5,6),align='left')

# The histogram gives an overview about the distribution of numerical data. Unlike the boxplot, the
# histogram doesn't give us the meta information like the mean, quartiles etc. but instead we get
# an exact amount for how many data points there are for a specific value. In a histogram, the values
# of the data can be ranked from small to big or similar. The bars touch their neighbours which shows
# the continous nature of the values -- for example you could see "The higher my values, the lower the
# amount of data points that have those values".

# While we may also be interested in knowing a little bit of meta information, for example the mean
# of all grades, for the most part we aren't very interested in it -- usually the median of all grades
# is not of importance, neither are quartiles. Also there are no real outliers within a grading system.
# Instead we want to know how many data points we have for each grade. This is exactly what a histogram
# presents us.
```

```python
# Get some data from your data folder

data = loadtxt('minimum_salary.dat')
 
plt.xlabel('country')
plt.ylabel('minimum salary in euros')
plt.title('Minimum salary')
plt.legend()

plt.bar(('BEL', 'GRE', 'SPA', 'FRA', 'LUX', 'NET', 'UK'), data)

# Although quite similar to a histogram, a bar chart is not exactly the same. A bar graph
# doesn't have numerical data to compare but instead uses different categories as separation.
# So unlike a histogram you can't rank the categories because there is no order, they are just
# different categories. To make this visible the bars in bar chart do not touch their
# neighbours. You thus can't get any information of the right bars being higher than the
# left bars because they might as well be ordered in a completely different way and the graph
# would still be the same.

# As in the previous answer for the histogram we again don't care for the meta information as
# median etc. What we want to know though is the minimum salary in each country and how they
# compare to one another. This time we don't need the amount of data points but instead want
# to know the value for each specific land. As the lands are not numerical but just different
# categories, we now use a bar chart for that.
```

```python
# Get some data from your data folder

data = loadtxt('music.dat')

plt.title('Favorite music styles')
plt.legend()

plt.pie(data,labels=('Grunge', 'Hip Hop', 'Metal', 'Schlager', 'Rock'))

# Being very similar to a bar chart a pie chart serves pretty much the exact same purpose.
# Again we want to compare values for different categories, although we could also have
# numerical data that would otherwise have been plotted within a histogram.
# It only makes sense to use a pie chart if you want to know how big each categorie's piece
# of the pie is. For this to make sense the values from all categories have to sum up to 
# something meaningful. Maybe you have a total of 100 % and each category has some of that
# percentage, or you have made a survey and have a total number of participants of which
# each category has some. For example, a pie chart wouldn't have made sense for the minimum
# salary data as the sum of all the minimum salaries doesn't make any sense.
# On the other hand, everything that can be plotted with a pie chart in a meaningful way can
# also be plotted with a bar chart (or histogram if the data is numerical). The advantage of
# a pie chart -- if the sum of all values makes sense and you want to know not the total
# number for each category but instead the percentage of the lot each category has -- is that
# this percentage can be visualized in a way that's easier to recognize and probably also
# more beautiful.

# As this data was actually from a survey the most interesting thing we want to know from the
# results is not the number of participants that voted for each music style but instead the
# percentage of the lot each category has. We want to see how big a category is compared to
# another. A pie chart presents this exact information in a beautiful way.
```
